Automatizando o processo de passar as variaveis da posição 0 como titulo. (Caso o banco de dados no exell tenha começado uma linha abaixo).

```python
index = 0
aux = []
for col in range(0, len(df.columns)):
  print(df.loc[0,f'Unnamed: {index}'])
  aux.append(df.loc[0,f'Unnamed: {index}'])
  index += 1
df.columns = aux
df
```

```python
df[:] = df[:].shift(-1) 
df = df.drop(df.index[-1])

df
```

```python
numero_de_anos = len(df.columns) - 1 
print(numero_de_anos)
aux1 = 2
index = 1
while index <= numero_de_anos:
  df.insert(aux1, f'X{index} Real', 0)
  df.insert(aux1+1, f'AH{index}', 0)
  aux1 += 3
  index += 1
# df.insert(2, 'X1 Real', 0)
# df.insert(3, 'AH1', 0)
# df.insert(5, 'X2 Real', 0)
# df.insert(6, 'AH2', 1)
# df.insert(8, 'X3 Real', 0)
# df.insert(9, 'AH3', 0)
# df
df
```

```python
indice1 = 1.2857
indice2 = 1.125
decimals = 2
i = 2
t = 1
df.iloc[:,i] = df[f'X{str(t)}']*indice1 # x1 real
df[f'X{t}'] = df[f'X{t}'].apply(lambda x: round(x, decimals))
df[f'X{t} Real'] = df[f'X{t} Real'].apply(lambda x: round(x, decimals))
df.iloc[:,i+1] = 100 # AH1
i += 3
t += 1
df.iloc[:,i] = df[f'X{str(t)}']*indice2
df.iloc[:,i+1] = (df[f'X{str(t)} Real']/df[f'X{str(t-1)} Real'])*100 # AH2
df[f'X{t} Real'] = df[f'X{t} Real'].apply(lambda x: round(x, decimals))
df[f'AH{t}'] = df[f'AH{t}'].apply(lambda x: round(x, decimals))
i += 3
t += 1
print(f'{t}')
print(f'{i}')
df.iloc[:,i] = df[f'X{str(t)}'] # x3 real
df.iloc[:,i+1] = (df[f'X{str(t)} Real']/df[f'X{str(1)} Real'])*100
df[f'X{t} Real'] = df[f'X{t} Real'].apply(lambda x: round(x, decimals))
df[f'AH{t}'] = df[f'AH{t}'].apply(lambda x: round(x, decimals)) 
i += 3
t += 1
df
```
